```python
# 전제: duckdb 설치 & con = duckdb.connect() 완료 상태
import os
import duckdb

file_path = "/content/drive/MyDrive/DILAB/MARS/mimiciv_3.1/files/hosp/pharmacy.csv"
out_path  = "/content/drive/MyDrive/DILAB/OK/DI_LAB/MARS_Datathon/Datasets/pharmacy_subject_10000032.csv"

# 출력 폴더 없으면 생성
os.makedirs(os.path.dirname(out_path), exist_ok=True)

# 모든 컬럼을 우선 문자열로 읽고( all_varchar=1 ), 파일 전체 스캔( sample_size=-1 )
# 문자열이 섞인 열(예: fill_quantity='q12h') 때문에 타입 추론이 깨지는 걸 방지
df = duckdb.connect().execute(f"""
WITH raw AS (
  SELECT *
  FROM read_csv_auto('{file_path}', all_varchar=1, sample_size=-1)
)
SELECT *
FROM raw
WHERE TRY_CAST(subject_id AS BIGINT) = 10000032
""").df()

# 저장
df.to_csv(out_path, index=False)

# 확인용 미리보기
df.head()
```

## subject_id=10000032인 모든 칼럼 합치는 코드

```python
import os
from glob import glob
import duckdb
import pandas as pd

# DuckDB 연결 (메모리)
con = duckdb.connect()

# 입력 폴더와 출력 파일
in_dir   = "/content/drive/MyDrive/DILAB/MARS/mimiciv_3.1/files/hosp"
out_path = "/content/drive/MyDrive/DILAB/OK/DI_LAB/MARS_Datathon/Datasets/subject_10000032_all.csv"
os.makedirs(os.path.dirname(out_path), exist_ok=True)

# 폴더 내 CSV 모두 대상
csv_files = sorted(glob(os.path.join(in_dir, "*.csv")))

frames = []
for fp in csv_files:
    # 1) 현재 파일의 컬럼 목록 살펴보기 (DuckDB에서 스키마 추론만)
    schema = con.execute(f"""
        DESCRIBE SELECT * FROM read_csv_auto('{fp}', all_varchar=1, sample_size=20480)
    """).df()
    cols_lower = set(schema["column_name"].str.lower().tolist())

    # 2) subject_id가 없는 파일(d_labitems 같은 사전테이블)은 스킵
    if "subject_id" not in cols_lower:
        # 필요하면 여기서 print(f"skip: {os.path.basename(fp)} (no subject_id)") 해도 됨
        continue

    # 3) subject_id=10000032만 필터해서 읽기 (전 컬럼 문자열 + 전체 스캔)
    #    "___" 같은 토큰을 결측으로 보고 싶으면 nullstr 옵션에 추가
    q = f"""
    WITH raw AS (
      SELECT *
      FROM read_csv_auto('{fp}', all_varchar=1, sample_size=-1, nullstr=['', 'NULL', 'NaN', '___'])
    )
    SELECT *, '{os.path.basename(fp)}' AS source_file
    FROM raw
    WHERE TRY_CAST(subject_id AS BIGINT) = 10000032
    """
    part = con.execute(q).df()

    if len(part):
        frames.append(part)

# 4) 파일별 결과를 '열 합집합'으로 세로 결합
if frames:
    combined = pd.concat(frames, axis=0, join="outer", ignore_index=True)
    combined.to_csv(out_path, index=False)
    print(f"완료! 행수={len(combined):,}  저장: {out_path}")
    display(combined.head(10))
else:
    print("subject_id=10000032 데이터가 포함된 파일이 없습니다.")
```
