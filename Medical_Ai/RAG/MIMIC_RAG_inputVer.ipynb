# 1) 모델 준비
/content/drive/MyDrive/DILAB/qwen3-8b

```python
MODEL_ID = "/content/drive/MyDrive/DILAB/qwen3-8b"

import torch
from transformers import AutoModelForCausalLM, AutoTokenizer

torch.backends.cuda.matmul.allow_tf32 = True
dtype = torch.bfloat16

tok = AutoTokenizer.from_pretrained(MODEL_ID, use_fast=True, trust_remote_code=True)
model = AutoModelForCausalLM.from_pretrained(
    MODEL_ID,
    torch_dtype=dtype,
    device_map="auto",
    trust_remote_code=True,
)

def generate_chat(messages, max_new_tokens=700):
    text = tok.apply_chat_template(messages, tokenize=False, add_generation_prompt=True)
    inputs = tok([text], return_tensors="pt").to(model.device)
    with torch.no_grad():
        out = model.generate(
            **inputs, max_new_tokens=max_new_tokens,
            do_sample=False, temperature=0.0, repetition_penalty=1.05
        )
    out_text = tok.decode(out[0], skip_special_tokens=True)
    return out_text.split("assistant\n")[-1].strip() if "assistant\n" in out_text else out_text
```

# 2) 입력 파켓 로드 -> 근거 본문 만들기
/content/drive/MyDrive/DILAB/MARS/mimic-iv-note_2.2/files/note/input_output/discharge_train_input.parquet

```python
import pandas as pd, numpy as np, re

INPUT_PATH = "/content/drive/MyDrive/DILAB/MARS/mimic-iv-note_2.2/files/note/input_output/discharge_train_input.parquet"  # 경로 변경 가능
df = pd.read_parquet(INPUT_PATH)

EXCLUDE_COLS = {"bhc","di"}  # 생성 대상은 제외
META_KEYS = ["SUBJECT_ID","HADM_ID","CHARTDATE","CHARTTIME"]

def clean_text(x):
    if x is None or (isinstance(x,float) and np.isnan(x)): return ""
    s = re.sub(r"\s+", " ", str(x)).strip()
    return s

def row_to_evidence_text(row, max_field_len=2000):
    parts=[]
    for col,val in row.items():
        if col.lower() in EXCLUDE_COLS:
            continue
        v = clean_text(val)
        if not v:
            continue
        if len(v)>max_field_len: v = v[:max_field_len] + " ..."
        parts.append(f"{col}: {v}")
    return "\n".join(parts)

# case_id: 같은 케이스만 검색하도록 메타 키 (우선순위: HADM_ID > SUBJECT_ID+CHARTDATE > 행번호)
def make_case_id(row):
    if "HADM_ID" in row and pd.notna(row["HADM_ID"]): return f"HADM:{row['HADM_ID']}"
    sid = str(row["SUBJECT_ID"]) if "SUBJECT_ID" in row and pd.notna(row["SUBJECT_ID"]) else None
    cdate = str(row["CHARTDATE"]) if "CHARTDATE" in row and pd.notna(row["CHARTDATE"]) else None
    if sid and cdate: return f"SID:{sid}_DATE:{str(cdate)[:10]}"
    return f"ROW:{int(row.name)}"
```

# 3) SemanticChunker로 의미 청크 생성

```python
from langchain_experimental.text_splitter import SemanticChunker
from langchain_community.embeddings import HuggingFaceEmbeddings
from langchain.schema import Document
from tqdm.auto import tqdm
from datetime import datetime

# 청커 내부 임베딩(LC 래퍼)
emb_lc = HuggingFaceEmbeddings(
    model_name="intfloat/e5-base",
    model_kwargs={"device":"cuda"},
    encode_kwargs={"normalize_embeddings": True},
)

chunker = SemanticChunker(
    emb_lc,
    breakpoint_threshold_type="percentile",
    breakpoint_threshold_amount=96,  # 높을수록 큰 덩어리
)

def build_documents_from_df(df, preview_every=50):
    """
    df 전 행을 의미 청크로 변환하면서:
      - tqdm 진행률 표시
      - preview_every 주기마다 간단 미리보기 출력
      - 누적 청크 수 / 실패 건수 실시간 갱신
    """
    docs = []
    meta_cols_present = [c for c in df.columns if c in META_KEYS]

    total_rows = len(df)
    total_chunks = 0
    failures = 0

    pbar = tqdm(df.iterrows(), total=total_rows, desc="Semantic chunking")
    for i, (_, row) in enumerate(pbar):
        try:
            text = row_to_evidence_text(row)
            case = make_case_id(row)

            base_meta = {k: row.get(k, None) for k in meta_cols_present}
            base_meta.update({"case_id": case, "_row_index": int(i)})

            chs = chunker.create_documents([text], metadatas=[base_meta])

            # doc_id 부여 + 누적
            for j, d in enumerate(chs):
                d.metadata["doc_id"] = f"case:{case}::chunk:{j}"
            docs.extend(chs)

            total_chunks += len(chs)

            # 상태 표시
            pbar.set_postfix(chunks=total_chunks, fails=failures)

            # 주기적 미리보기
            if preview_every and ((i + 1) % preview_every == 0 or (i + 1) == total_rows):
                sample = (chs[0].page_content if chs else "")[:160].replace("\n", " ")
                tqdm.write(
                    f"[{datetime.now().strftime('%H:%M:%S')}] "
                    f"row {i+1}/{total_rows} | case={case} | +{len(chs)} chunks "
                    f"(sum={total_chunks}) | sample: {sample}"
                )

        except Exception as e:
            failures += 1
            pbar.set_postfix(chunks=total_chunks, fails=failures)
            tqdm.write(f"[WARN] row {i+1} failed: {e}")

    tqdm.write(f"Done. rows={total_rows}, chunks={total_chunks}, failures={failures}")
    return docs

documents = build_documents_from_df(df)
len(documents)
```

# 4) 벡터 DB(FAISS) 구축 & 저장

```python
from langchain_community.vectorstores import FAISS
from tqdm.auto import tqdm
from datetime import datetime
import os

def build_faiss_with_progress(documents, emb, save_dir,
                              index_name="faiss_inputs_semantic",
                              batch_size=256, preview_every=10):
    """
    documents: list[Document]
    emb: HuggingFaceEmbeddings (or 호환)
    save_dir: 저장할 베이스 디렉터리
    index_name: save_local에 쓸 폴더명
    batch_size: 임베딩/추가 배치 크기
    preview_every: N배치마다 샘플 출력 (0이면 미리보기 끔)
    """
    os.makedirs(save_dir, exist_ok=True)

    total = len(documents)
    if total == 0:
        raise ValueError("documents가 비어 있습니다.")

    vectorstore = None
    added = 0
    num_batches = (total + batch_size - 1) // batch_size
    pbar = tqdm(range(0, total, batch_size), total=num_batches, desc="Building FAISS")

    for b_start in pbar:
        batch = documents[b_start:b_start+batch_size]
        if vectorstore is None:
            # 첫 배치는 from_documents로 초기화
            vectorstore = FAISS.from_documents(batch, emb)
            added += len(batch)
        else:
            # 이후 배치는 add_documents로 증분 추가
            vectorstore.add_documents(batch)
            added += len(batch)

        pbar.set_postfix(added=added)

        # 주기적 미리보기
        if preview_every and (((b_start // batch_size + 1) % preview_every) == 0 or added == total):
            d0 = batch[0]
            sid = d0.metadata.get("SUBJECT_ID", "")
            hadm = d0.metadata.get("HADM_ID", "")
            cid = d0.metadata.get("case_id", "")
            sample = d0.page_content[:160].replace("\n", " ")
            tqdm.write(
                f"[{datetime.now().strftime('%H:%M:%S')}] "
                f"added {added}/{total} | case={cid} | SUBJECT_ID={sid} HADM_ID={hadm} | sample: {sample}"
            )

    # 저장
    save_path = os.path.join(save_dir, index_name)
    vectorstore.save_local(save_path)
    tqdm.write(f"Saved FAISS to: {save_path}")

    return vectorstore

# 사용 예시
save_base = "/content/drive/MyDrive/DILAB/OK_/rag_store"
vectorstore = build_faiss_with_progress(
    documents,
    emb_lc,
    save_dir=save_base,
    index_name="faiss_inputs_semantic",
    batch_size=256,      # A100이면 256~512도 OK (OOM 시 줄이기)
    preview_every=5      # 5배치마다 미리보기
)
```

# 5) 검색 + 생성 함수 (전역 인덱스에서 “같은 케이스만” 필터)

```python
SYSTEM_PROMPT = (
    "You are a clinical documentation assistant. Answer in English, grounded strictly in the provided evidence. "
    "If unknown, write 'Not documented'. Return valid JSON with keys 'bhc' and 'di'."
)

def retrieve_for_row(row, k=8, fetch_k=64):
    """전역 인덱스에서 우선 fetch_k개를 가져온 뒤, 같은 case_id만 남겨 상위 k개로 축소."""
    case = make_case_id(row)
    query = (row_to_evidence_text(row)[:2000] + "\n\nFocus: brief hospital course and discharge instructions").strip()
    # 1) 넉넉히 검색
    hits = vectorstore.similarity_search_with_score(query, k=fetch_k)
    # 2) 같은 케이스만 필터
    same = [(d,s) for d,s in hits if d.metadata.get("case_id")==case]
    if not same:  # 안전망: 그래도 없으면 원래 hits에서 상위 k 사용(이 경우 교차오염 가능성, 로그 남김)
        same = hits
    # 3) 상위 k로 제한
    same = same[:k]
    return same  # [(Document, score), ...]

def build_context(ranked, max_chars=4500):
    parts=[]
    for d, sc in ranked:
        md = d.metadata or {}
        head = " | ".join([f"{k}={md[k]}" for k in ["case_id","SUBJECT_ID","HADM_ID","CHARTDATE"] if k in md and md[k] is not None])
        parts.append(f"[{head}] (score={sc:.3f})\n{d.page_content}")
    ctx = "\n\n---\n\n".join(parts)
    return ctx[:max_chars]

import json, textwrap

def generate_bhc_di_from_row(row, k=8, fetch_k=64):
    ranked = retrieve_for_row(row, k=k, fetch_k=fetch_k)
    context = build_context(ranked)
    user_prompt = textwrap.dedent(f"""
    Use ONLY the evidence below to draft two discharge summary sections.

    ## Evidence
    {context}

    ## Task
    Return a JSON object with:
    - "bhc": Brief Hospital Course — chronological; major diagnoses, key diagnostics, interventions/treatments and response; 440-550 words.
    - "di": Discharge Instructions — 5–10 bullet points on meds (incl. changes), activity, diet, follow-up, warning signs, patient education.

    ## Rules
    - English only.
    - If a fact is not documented, write "Not documented".
    - Do not include PHI.
    """).strip()

    messages = [
        {"role":"system","content":SYSTEM_PROMPT},
        {"role":"user","content":user_prompt}
    ]
    raw = generate_chat(messages)
    try:
        s, e = raw.find("{"), raw.rfind("}")
        obj = json.loads(raw[s:e+1])
    except Exception:
        obj = {"bhc": raw, "di":"Not documented"}
    bhc = (obj.get("bhc") or "Not documented").strip()
    di  = (obj.get("di")  or "Not documented").strip()
    used = [d.metadata.get("doc_id") for d,_ in ranked]
    return bhc, di, used
```

## 5-1) 수정버전 -> 실패(에러뜸)

```python
# 수정 코드
# === 리트리버: 동일 케이스만 유지 + 넉넉히 가져와서 MMR로 k개 선택 ===
import numpy as np
import re, json, torch

def _mmr_select(query_emb, doc_embs, k=8, lambda_=0.7):
    """
    간단 MMR(Maximal Marginal Relevance) 구현.
    query_emb: (d,)
    doc_embs:  (N, d) - normalize되어 있으면 코사인=내적
    """
    if len(doc_embs) == 0:
        return []
    sims_to_query = doc_embs @ query_emb  # (N,)
    selected = []
    candidates = list(range(len(doc_embs)))
    # 1개는 최고 유사도부터
    first = int(np.argmax(sims_to_query))
    selected.append(first)
    candidates.remove(first)

    while len(selected) < min(k, len(doc_embs)) and candidates:
        mmr_scores = []
        for c in candidates:
            sim_to_q = sims_to_query[c]
            if selected:
                sim_to_sel = np.max(doc_embs[c] @ doc_embs[selected].T) if len(selected) > 1 else (doc_embs[c] @ doc_embs[selected[0]].T)
            else:
                sim_to_sel = 0.0
            mmr = lambda_ * sim_to_q - (1 - lambda_) * sim_to_sel
            mmr_scores.append((mmr, c))
        _, best = max(mmr_scores, key=lambda x: x[0])
        selected.append(best)
        candidates.remove(best)
    return selected

def retrieve_for_row_strict(row, k=8, fetch_k=64, strict=True):
    """
    전역 FAISS에서 우선 fetch_k개 검색 → 같은 case_id만 필터(하드) → MMR로 k개 선정.
    strict=True면, 같은 case_id 히트가 없을 때 '빈 컨텍스트' 반환(타 케이스 폴백 금지).
    """
    case = make_case_id(row)
    query = (row_to_evidence_text(row)[:2000] +
             "\n\nFocus: brief hospital course and discharge instructions").strip()

    # 1) 넉넉히 가져오기
    hits = vectorstore.similarity_search_with_score(query, k=fetch_k)

    # 2) 같은 케이스만 하드 필터
    same = [(d, s) for d, s in hits if d.metadata.get("case_id") == case]
    if strict and not same:
        return []  # 타 케이스 사용 금지

    pool = same if same else hits  # strict=False일 때만 타 케이스 허용

    # 3) MMR로 k개 고르기
    cand_texts = [d.page_content for d, _ in pool]
    # HuggingFaceEmbeddings는 normalize 옵션을 켠 상태(권장). 쿼리/문서 둘 다 정규화됨.
    q_emb = np.array(emb_lc.embed_query(query), dtype="float32")
    doc_embs = np.array(emb_lc.embed_documents(cand_texts), dtype="float32")
    sel_idx = _mmr_select(q_emb, doc_embs, k=k, lambda_=0.7)

    ranked = [(pool[i][0], float(pool[i][1])) for i in sel_idx]
    return ranked

def build_context(ranked, max_chars=6000):
    parts = []
    for d, sc in ranked:
        md = d.metadata or {}
        head = " | ".join([f"{k}={md[k]}" for k in ["case_id","SUBJECT_ID","HADM_ID","CHARTDATE"] if k in md and md[k] is not None])
        parts.append(f"[{head}] (score={sc:.3f})\n{d.page_content}")
    ctx = "\n\n---\n\n".join(parts)
    return ctx[:max_chars]

SYSTEM_PROMPT = (
    "You are a clinical documentation assistant. Answer in English.\n"
    "Use ONLY facts explicitly present in EVIDENCE; do NOT add labs, doses, volumes, procedures, or dates that are not literally present.\n"
    "If a fact is missing, write 'Not documented'. Return VALID JSON only. Do not reveal your reasoning."
)

def generate_json_only(messages, max_new_tokens=700):
    """모델 출력을 JSON 블록만 추출. 실패 시 1회 재지시 후 재시도."""
    text = tok.apply_chat_template(messages, tokenize=False, add_generation_prompt=True)
    inputs = tok([text], return_tensors="pt").to(model.device)
    with torch.no_grad():
        out = model.generate(**inputs, max_new_tokens=max_new_tokens, do_sample=False, temperature=0.0, repetition_penalty=1.05)
    raw = tok.decode(out[0], skip_special_tokens=True)
    raw = raw.split("assistant\n")[-1].strip()

    m = re.search(r"\{[\s\S]*\}\s*$", raw)
    if m:
        return json.loads(m.group(0))

    # 1회만 강제 재시도
    messages2 = messages[:-1] + [{"role":"user","content": messages[-1]["content"] + "\n\nReturn ONLY a valid JSON object now."}]
    text2 = tok.apply_chat_template(messages2, tokenize=False, add_generation_prompt=True)
    inputs2 = tok([text2], return_tensors="pt").to(model.device)
    with torch.no_grad():
        out2 = model.generate(**inputs2, max_new_tokens=400, do_sample=False, temperature=0.0)
    raw2 = tok.decode(out2[0], skip_special_tokens=True)
    m2 = re.search(r"\{[\s\S]*\}\s*$", raw2)
    if not m2:
        # 최후 폴백
        return {"bhc":"Not documented", "di":"Not documented"}
    return json.loads(m2.group(0))

    # 숫자 검출
def _extract_numbers(s):
    return set(re.findall(r"\b\d+(?:\.\d+)?\b", s))

# 증거에 없는 특정 용어(시술/약물 등) 문장 제거/치환
_UNSUPPORTED_TERMS = [
    "paracentesis", "albumin", "midodrine", "octreotide", "rifaximin",
    "lactulose", "TIPS", "MELD", "Child", "EGD", "colonoscopy", "Bactrim",
]

def _mask_unsupported_terms(text, evidence):
    ev_low = evidence.lower()
    out = text
    for t in _UNSUPPORTED_TERMS:
        if t.lower() not in ev_low:
            # 해당 단어가 들어간 문장 통째로 'Not documented.'로 치환
            pattern = rf"[^.\n]*\b{re.escape(t)}\b[^.\n]*[.\n]"
            out = re.sub(pattern, "Not documented.\n", out, flags=re.IGNORECASE)
    return out

def _enforce_word_count(bhc_text, min_w=120, max_w=220):
    words = re.findall(r"\w+", bhc_text)
    return len(words), (len(words) >= min_w and len(words) <= max_w)

def _count_bullets(di_text):
    lines = [l.strip() for l in di_text.splitlines()]
    bullets = [l for l in lines if re.match(r"^(\-|\*|•)\s+", l)]
    return len(bullets)

def _join_bullets(items):
    items = [f"- {it.strip()}" for it in items if it and it.strip()]
    return "\n".join(items)
```

```python
import textwrap

def generate_bhc_di_from_row(row, k=8, fetch_k=64):
    # 1) 검색
    ranked = retrieve_for_row_strict(row, k=k, fetch_k=fetch_k, strict=True)
    context = build_context(ranked)

    # 2) 1차 생성(JSON-only)
    user_prompt = textwrap.dedent(f"""
    EVIDENCE:
    {context}

    TASK:
    Return a JSON object with exactly two keys:
    - "bhc": 120–220 words, chronological hospital course. Copy numbers, doses, volumes, medication names verbatim from EVIDENCE.
    - "di": 5–10 bullet points; meds/doses, diet, activity, follow-up, warning signs, education. Copy numbers/med names verbatim.

    RULES:
    - If a fact (number/lab/dose/procedure/date) is not literally present in EVIDENCE, write "Not documented".
    - Do not reveal chain-of-thought. JSON ONLY.
    """).strip()

    messages = [
        {"role":"system","content": SYSTEM_PROMPT},
        {"role":"user","content": user_prompt},
    ]
    obj = generate_json_only(messages)

    bhc = (obj.get("bhc") or "Not documented").strip()
    di  = (obj.get("di")  or "Not documented").strip()

    # 3) 사후 검증(숫자/키워드)
    ev = context  # evidence 텍스트
    gen_nums = _extract_numbers(bhc + " " + di)
    ev_nums  = _extract_numbers(ev)
    illegal_nums = gen_nums - ev_nums
    if illegal_nums:
        # 숫자 환각이 있으면 숫자 전체를 'Not documented'로 치환
        bhc = re.sub(r"\b\d+(\.\d+)?\b", lambda m: m.group(0) if m.group(0) in ev_nums else "Not documented", bhc)
        di  = re.sub(r"\b\d+(\.\d+)?\b",  lambda m: m.group(0) if m.group(0) in ev_nums else "Not documented", di)

    # 증거에 없는 주요 용어 문장 제거/치환
    bhc = _mask_unsupported_terms(bhc, ev)
    di  = _mask_unsupported_terms(di, ev)

    # 4) 형식 검증: BHC 단어수, DI 불릿수
    wc, ok_wc = _enforce_word_count(bhc, 120, 220)
    bullets = _count_bullets(di)
    need_fix_bhc = not ok_wc
    need_fix_di  = not (5 <= bullets <= 10)

    # 4-1) BHC 길이 보정(재지시)
    if need_fix_bhc:
        fix_prompt = textwrap.dedent(f"""
        EVIDENCE:
        {ev}

        TASK:
        Rewrite "bhc" ONLY as a single paragraph of 150–200 words (strict), chronological, copying numbers/meds verbatim from EVIDENCE. If missing, write "Not documented".
        Return JSON with only the key "bhc".
        """).strip()
        obj2 = generate_json_only([{"role":"system","content": SYSTEM_PROMPT},
                                   {"role":"user","content": fix_prompt}], max_new_tokens=350)
        bhc = (obj2.get("bhc") or bhc).strip()
        # 숫자/용어 재검사
        gen_nums = _extract_numbers(bhc)
        illegal  = gen_nums - ev_nums
        if illegal:
            bhc = re.sub(r"\b\d+(\.\d+)?\b", lambda m: m.group(0) if m.group(0) in ev_nums else "Not documented", bhc)
        bhc = _mask_unsupported_terms(bhc, ev)

    # 4-2) DI 불릿 보정(재지시)
    if need_fix_di or di.lower().strip() in ["", "not documented"]:
        fix_prompt = textwrap.dedent(f"""
        EVIDENCE:
        {ev}

        TASK:
        Return JSON with only the key "di" whose value is an array (list) of 5–10 concise bullet strings.
        Each bullet MUST copy any numbers/med names verbatim from EVIDENCE (otherwise write "Not documented").
        """).strip()
        obj3 = generate_json_only([{"role":"system","content": SYSTEM_PROMPT},
                                   {"role":"user","content": fix_prompt}], max_new_tokens=350)
        di_list = obj3.get("di")
        if isinstance(di_list, list) and 5 <= len(di_list) <= 10:
            di = _join_bullets(di_list)
            # 숫자/용어 재검사
            gen_nums = _extract_numbers(di)
            illegal  = gen_nums - ev_nums
            if illegal:
                di = re.sub(r"\b\d+(\.\d+)?\b", lambda m: m.group(0) if m.group(0) in ev_nums else "Not documented", di)
            di = _mask_unsupported_terms(di, ev)

    used = [d.metadata.get("doc_id") for d,_ in ranked]
    return bhc, di, used
```

```python
# ==== 준비 셀 (df 없이): emb_lc, vectorstore, tok/model 로드 ====
import os, torch
from langchain_community.embeddings import HuggingFaceEmbeddings
from langchain_community.vectorstores import FAISS
from transformers import AutoTokenizer, AutoModelForCausalLM

# 1) 임베딩 (인덱스 빌드 때와 동일 모델/옵션 권장)
EMB_NAME = "intfloat/e5-base"  # ← 인덱스 구축 때 쓴 모델명으로 맞추세요
emb_lc = HuggingFaceEmbeddings(
    model_name=EMB_NAME,
    model_kwargs={"device":"cuda" if torch.cuda.is_available() else "cpu"},
    encode_kwargs={"normalize_embeddings": True},
)
E5_QUERY_PREFIX = "query: "   # e5 계열 쿼리 접두어 권장
print(f"[init] emb_lc ready: {EMB_NAME}")

# 2) 기존 FAISS 인덱스 로드 (폴더 안에 index.faiss + index.pkl 둘 다 있어야 함)
def _load_faiss_or_discover(root_list, index_name, emb):
    import glob
    for root in root_list:
        path = os.path.join(root, index_name)
        faissf = os.path.join(path, "index.faiss")
        pklf   = os.path.join(path, "index.pkl")
        if os.path.isdir(path) and os.path.exists(faissf) and os.path.exists(pklf):
            print(f"[load] using: {path}")
            return FAISS.load_local(path, emb, allow_dangerous_deserialization=True)
    # 재귀 탐색(다른 폴더에 저장됐을 가능성)
    for root in root_list:
        for p in glob.glob(os.path.join(root, "**", "index.faiss"), recursive=True):
            folder = os.path.dirname(p)
            if os.path.exists(os.path.join(folder, "index.pkl")):
                print(f"[load] discovered: {folder}")
                return FAISS.load_local(folder, emb, allow_dangerous_deserialization=True)
    raise FileNotFoundError("index.faiss/index.pkl 세트를 찾지 못했습니다. 루트/폴더명을 확인하세요.")

roots = [
    "/content/drive/MyDrive/DILAB/OK_Data/rag_store",  # 지금 쓰는 경로
    "/content/drive/MyDrive/DILAB/OK_/rag_store",      # 과거에 썼을 수 있는 경로
]
vectorstore = _load_faiss_or_discover(roots, "faiss_inputs_semantic", emb_lc)
try:
    ntotal = getattr(getattr(vectorstore, "index", None), "ntotal", None)
    print(f"[init] vectorstore loaded. ntotal={ntotal}")
except Exception:
    print("[init] vectorstore loaded.")
```

```python
# === Retriever & Generator (증거 기반 JSON 전용 파이프라인) ===
import os
import numpy as np
import re, json, torch
from typing import List
from langchain_community.vectorstores import FAISS
from tqdm import tqdm

# -------------------------------------------------
# (추가) 케이스ID/쿼리텍스트 헬퍼  ─ df 없이 동작
# -------------------------------------------------
_TEXT_CANDS = ["TEXT","text","note_text","note","content","CONTENT","body"]

def _detect_content_col_from_row(row) -> str | None:
    keys = list(getattr(row, "index", getattr(row, "keys", lambda: [])()) or [])
    for c in _TEXT_CANDS:
        if c in keys:
            return c
    df_ref = globals().get("df", None)  # 있으면 보조로 사용(없어도 OK)
    if df_ref is not None:
        for c in _TEXT_CANDS:
            if c in df_ref.columns:
                return c
    return None

def make_case_id(row) -> str:
    sid  = str(row.get("SUBJECT_ID", row.get("subject_id","")))
    hadm = str(row.get("HADM_ID",  row.get("hadm_id","")))
    return f"{sid}-{hadm}"

def row_to_evidence_text(row) -> str:
    bits = []
    for k in ["CATEGORY","TITLE","DESCRIPTION","CHARTDATE"]:
        if (k in row) and (row[k] is not None) and str(row[k]).strip():
            bits.append(f"{k}: {row[k]}")
    content_col = _detect_content_col_from_row(row)
    body = str(row.get(content_col, "Not documented.")) if content_col else "Not documented."
    return ("\n".join(bits) + ("\n\n" if bits else "") + body).strip()

# -------------------------------------------------
# A) (선택) FAISS 빌더 구현: 필요 시에만 사용. 지금은 로드만 할 예정.
# -------------------------------------------------
def build_faiss_with_progress(
    documents,
    emb_lc,
    save_dir: str,
    index_name: str = "faiss_inputs_semantic",
    batch_size: int = 256,
    preview_every: int = 5,
):
    if not documents:
        raise ValueError("documents 가 비어 있습니다.")
    if emb_lc is None:
        raise ValueError("emb_lc 가 필요합니다.")
    vs = None
    for i in tqdm(range(0, len(documents), batch_size), desc="Build FAISS"):
        chunk = documents[i:i+batch_size]
        if vs is None:
            vs = FAISS.from_documents(chunk, emb_lc)
        else:
            vs.add_documents(chunk)
        if preview_every and (i // batch_size) % max(1, preview_every) == 0:
            print(f"- indexed: {i+len(chunk)}/{len(documents)}")
    os.makedirs(save_dir, exist_ok=True)
    vs.save_local(os.path.join(save_dir, index_name))
    print(f"FAISS saved to: {os.path.join(save_dir, index_name)}")
    return vs

# -------------------------------------------------
# B) 출력 정규화 유틸
# -------------------------------------------------
_THINK_RE = re.compile(r"<think>[\s\S]*?</think>\s*", flags=re.IGNORECASE)
_JSON_TAIL_RE = re.compile(r"\{[\s\S]*\}\s*$")

def _strip_think_blocks(text: str) -> str:
    return _THINK_RE.sub("", text)

def _extract_json_tail(text: str) -> str | None:
    m = _JSON_TAIL_RE.search(text.strip())
    return m.group(0) if m else None

def _safe_json_loads(s: str) -> dict | None:
    try:
        return json.loads(s)
    except Exception:
        return None

def _last_nonempty_assistant(raw: str) -> str:
    return raw.split("assistant\n")[-1].strip()

# -------------------------------------------------
# C) MMR
# -------------------------------------------------
def _mmr_select(query_emb: np.ndarray, doc_embs: np.ndarray, k: int = 8, lambda_: float = 0.7) -> List[int]:
    if doc_embs is None or len(doc_embs) == 0:
        return []
    N = doc_embs.shape[0]
    k = max(1, min(k, N))
    sims_to_query = doc_embs @ query_emb
    selected = []
    candidates = list(range(N))
    first = int(np.argmax(sims_to_query))
    selected.append(first); candidates.remove(first)
    while len(selected) < k and candidates:
        best_score, best_idx = None, None
        sel_mat = doc_embs[selected]
        for c in candidates:
            sim_to_q = float(sims_to_query[c])
            sim_to_sel = float(np.max(doc_embs[c] @ sel_mat.T)) if sel_mat.ndim == 2 else float(doc_embs[c] @ sel_mat.T)
            mmr = lambda_ * sim_to_q - (1 - lambda_) * sim_to_sel
            if (best_score is None) or (mmr > best_score):
                best_score, best_idx = mmr, c
        selected.append(best_idx); candidates.remove(best_idx)
    seen, dedup = set(), []
    for i in selected:
        if i not in seen:
            dedup.append(i); seen.add(i)
    return dedup

# -------------------------------------------------
# D) 검색 단계 (e5 쿼리 접두어 반영)
# -------------------------------------------------
def retrieve_for_row_strict(row, k: int = 8, fetch_k: int = 64, strict: bool = True):
    """
    전역 vectorstore 에서 fetch_k 검색 → 같은 case_id만 필터 → MMR로 k 선택.
    """
    case = make_case_id(row)
    query = (row_to_evidence_text(row)[:2000] +
             "\n\nFocus: brief hospital course and discharge instructions").strip()

    hits = vectorstore.similarity_search_with_score(query, k=fetch_k)

    same = [(d, s) for d, s in hits if d.metadata.get("case_id") == case]
    if strict and not same:
        return []

    pool = same if same else hits
    cand_texts = [d.page_content for d, _ in pool]
    try:
        # e5 권장 접두어 (준비 셀에서 정의했다면 그 값을 사용, 없으면 공백)
        prefix = globals().get("E5_QUERY_PREFIX", "")
        q_emb = np.array(emb_lc.embed_query(prefix + query), dtype="float32")
        doc_embs = np.array(emb_lc.embed_documents(cand_texts), dtype="float32")
        if q_emb.ndim != 1:
            q_emb = q_emb.squeeze()
        sel_idx = _mmr_select(q_emb, doc_embs, k=k, lambda_=0.7)
    except Exception:
        sel_idx = list(range(min(k, len(pool))))

    ranked = [(pool[i][0], float(pool[i][1])) for i in sel_idx]
    return ranked

def build_context(ranked, max_chars: int = 6000) -> str:
    parts = []
    for d, sc in ranked:
        md = d.metadata or {}
        head = " | ".join([f"{k}={md[k]}" for k in ["case_id","SUBJECT_ID","HADM_ID","CHARTDATE"] if k in md and md[k] is not None])
        parts.append(f"[{head}] (score={sc:.3f})\n{d.page_content}")
    ctx = "\n\n---\n\n".join(parts)
    return ctx[:max_chars]

# -------------------------------------------------
# E) 생성 지시문
# -------------------------------------------------
SYSTEM_PROMPT = (
    "You are a clinical documentation assistant. Answer in English.\n"
    "Use ONLY facts explicitly present in EVIDENCE; do NOT add labs, doses, volumes, procedures, or dates that are not literally present.\n"
    "If a fact is missing, write 'Not documented'. Return VALID JSON only. Do not reveal your reasoning."
)

GEN_USER_INSTRUCTION = (
    "Using ONLY the EVIDENCE below, draft two discharge summary sections:\n"
    "- bhc: Brief Hospital Course in 120–220 words, concise paragraph(s).\n"
    "- di: Discharge Instructions as 3–7 bullet points.\n"
    "Do not add facts not literally present. If absent, state 'Not documented'.\n"
    "Return only a valid JSON object with keys 'bhc' and 'di'."
)

# -------------------------------------------------
# F) 사후 정제
# -------------------------------------------------
_UNSUPPORTED_TERMS = [
    "paracentesis", "albumin", "midodrine", "octreotide", "rifaximin",
    "lactulose", "TIPS", "MELD", "Child", "EGD", "colonoscopy", "Bactrim",
]

def _extract_numbers(s: str):
    return set(re.findall(r"\b\d+(?:\.\d+)?\b", s))

def _mask_unsupported_terms(text: str, evidence: str) -> str:
    ev_low = evidence.lower()
    out = text
    for t in _UNSUPPORTED_TERMS:
        if t.lower() not in ev_low:
            pattern = rf"[^.\n]*\b{re.escape(t)}\b[^.\n]*[.\n]"
            out = re.sub(pattern, "Not documented.\n", out, flags=re.IGNORECASE)
    return out

def _mask_new_numbers(text: str, evidence: str) -> str:
    ev_nums = _extract_numbers(evidence)
    def _scrub_line(line: str) -> str:
        nums = _extract_numbers(line)
        if any(n not in ev_nums for n in nums):
            return "Not documented."
        return line
    lines = re.split(r"(?<=[.\n])", text)
    lines = [_scrub_line(ln) for ln in lines]
    return "".join(lines)

def _enforce_bhc_length(bhc_text: str, min_w: int = 120, max_w: int = 220) -> str:
    words = re.findall(r"\w+[\-\w]*", bhc_text)
    n = len(words)
    if n < min_w:
        pad = " Not documented."
        while len(re.findall(r"\w+[\-\w]*", bhc_text)) < min_w:
            bhc_text += pad
    elif n > max_w:
        keep = words[:max_w]
        bhc_text = " ".join(keep).rstrip(".") + "."
    return bhc_text

def _normalize_bullets(di_text: str) -> str:
    items = []
    for line in di_text.splitlines():
        m = re.match(r"^\s*(?:\-|\*|•)\s+(.*\S)\s*$", line)
        if m:
            items.append(m.group(1).strip())
    if not items:
        sentences = [s.strip() for s in re.split(r"[.\n]+", di_text) if s.strip()]
        items = sentences[:5] if sentences else ["Not documented."]*3
    if len(items) < 3:
        items += ["Not documented."] * (3 - len(items))
    if len(items) > 7:
        items = items[:7]
    return "\n".join(f"- {it}" for it in items)

def _postprocess_to_evidence(json_obj: dict, evidence: str) -> dict:
    bhc = json_obj.get("bhc", "") or ""
    di  = json_obj.get("di", "") or ""

    bhc = _mask_unsupported_terms(bhc, evidence)
    bhc = _mask_new_numbers(bhc, evidence)
    di  = _mask_unsupported_terms(di, evidence)
    di  = _mask_new_numbers(di, evidence)

    bhc = _enforce_bhc_length(bhc, 120, 220)
    di  = _normalize_bullets(di)

    return {"bhc": bhc.strip(), "di": di.strip()}

# -------------------------------------------------
# G) 생성 호출부
# -------------------------------------------------
def generate_json_only(messages, max_new_tokens: int = 700) -> dict:
    if ('tok' not in globals()) or ('model' not in globals()):
        raise RuntimeError("tok/model 이 준비되어 있어야 합니다. (준비 셀에서 로드)")
    text = tok.apply_chat_template(messages, tokenize=False, add_generation_prompt=True)
    inputs = tok([text], return_tensors="pt").to(model.device)
    with torch.no_grad():
        out = model.generate(
            **inputs,
            max_new_tokens=max_new_tokens,
            do_sample=False,
            temperature=0.0,
            repetition_penalty=1.05,
        )
    raw = tok.decode(out[0], skip_special_tokens=True)
    raw = _last_nonempty_assistant(_strip_think_blocks(raw))

    js = _extract_json_tail(raw)
    if js:
        obj = _safe_json_loads(js)
        if obj and isinstance(obj, dict) and ("bhc" in obj and "di" in obj):
            return obj

    messages2 = messages[:-1] + [{
        "role":"user",
        "content": messages[-1]["content"] + "\n\nReturn ONLY a valid JSON object now."
    }]
    text2 = tok.apply_chat_template(messages2, tokenize=False, add_generation_prompt=True)
    inputs2 = tok([text2], return_tensors="pt").to(model.device)
    with torch.no_grad():
        out2 = model.generate(
            **inputs2,
            max_new_tokens=400,
            do_sample=False,
            temperature=0.0,
        )
    raw2 = tok.decode(out2[0], skip_special_tokens=True)
    raw2 = _last_nonempty_assistant(_strip_think_blocks(raw2))
    js2 = _extract_json_tail(raw2)
    if js2:
        obj2 = _safe_json_loads(js2)
        if obj2 and isinstance(obj2, dict) and ("bhc" in obj2 and "di" in obj2):
            return obj2
    return {"bhc":"Not documented", "di":"Not documented"}

# -------------------------------------------------
# H) row -> JSON
# -------------------------------------------------
def draft_sections(row, k: int = 8, fetch_k: int = 64, strict: bool = True, max_new_tokens: int = 700) -> dict:
    ranked = retrieve_for_row_strict(row, k=k, fetch_k=fetch_k, strict=strict)
    if not ranked:
        return {"bhc":"Not documented", "di":"Not documented"}
    evidence = build_context(ranked, max_chars=6000)
    messages = [
        {"role":"system","content": SYSTEM_PROMPT},
        {"role":"user","content": f"EVIDENCE:\n{evidence}\n\n{GEN_USER_INSTRUCTION}"}
    ]
    raw_obj = generate_json_only(messages, max_new_tokens=max_new_tokens)
    final_obj = _postprocess_to_evidence(raw_obj, evidence)
    return final_obj

# -------------------------------------------------
# I) FAISS 로드 (경로 자동 탐색 포함)
# -------------------------------------------------
save_base = "/content/drive/MyDrive/DILAB/OK_Data/rag_store"
alt_base  = "/content/drive/MyDrive/DILAB/OK_/rag_store"  # 과거 경로일 수 있음
index_name = "faiss_inputs_semantic"

def _load_faiss_or_discover(root_list, index_name, emb):
    import glob
    for root in root_list:
        path = os.path.join(root, index_name)
        if os.path.isdir(path) and os.path.exists(os.path.join(path,"index.faiss")) and os.path.exists(os.path.join(path,"index.pkl")):
            print(f"[load] using: {path}")
            return FAISS.load_local(path, emb, allow_dangerous_deserialization=True)
    for root in root_list:
        for p in glob.glob(os.path.join(root, "**", "index.faiss"), recursive=True):
            folder = os.path.dirname(p)
            if os.path.exists(os.path.join(folder,"index.pkl")):
                print(f"[load] discovered: {folder}")
                return FAISS.load_local(folder, emb, allow_dangerous_deserialization=True)
    raise FileNotFoundError(f"index.faiss/index.pkl 세트를 찾지 못했습니다. roots={root_list}, index_name={index_name}")

if 'emb_lc' not in globals():
    raise RuntimeError("emb_lc (임베딩 모델)가 먼저 준비되어 있어야 합니다. 준비 셀을 실행하세요.")

vectorstore = _load_faiss_or_discover([save_base, alt_base], index_name, emb_lc)

```

## 6-1) subjectId와 hadamId, noteId직접 지정

```python
# ---------------------------
# (수정된) 케이스ID/쿼리텍스트 헬퍼
# ---------------------------
_TEXT_CANDS = ["TEXT","text","note_text","note","content","CONTENT","body"]

def _detect_content_col_from_row(row) -> str | None:
    """row 자체에서 텍스트 컬럼명을 추정"""
    # pandas Series면 .index, dict-like면 keys 사용
    keys = list(getattr(row, "index", getattr(row, "keys", lambda: [])()) or [])
    for c in _TEXT_CANDS:
        if c in keys:
            return c
    # 최후 보조: 전역 df가 있으면 df.columns에서 추정 (없어도 통과)
    df_ref = globals().get("df", None)
    if df_ref is not None:
        for c in _TEXT_CANDS:
            if c in df_ref.columns:
                return c
    return None

def make_case_id(row) -> str:
    """동일 케이스 판별용 ID (기본: SUBJECT_ID-HADM_ID)"""
    sid  = str(row.get("SUBJECT_ID", row.get("subject_id","")))
    hadm = str(row.get("HADM_ID",  row.get("hadm_id","")))
    return f"{sid}-{hadm}"

def row_to_evidence_text(row) -> str:
    """검색 질의에 쓸 텍스트 생성 (df에 의존하지 않음)"""
    bits = []
    for k in ["CATEGORY","TITLE","DESCRIPTION","CHARTDATE"]:
        if (k in row) and pd.notnull(row[k]):
            bits.append(f"{k}: {row[k]}")
    content_col = _detect_content_col_from_row(row)
    if content_col is None:
        # 텍스트 본문을 찾지 못한 경우 안전 메시지
        body = "Not documented."
    else:
        body = str(row.get(content_col, ""))
    return ("\n".join(bits) + ("\n\n" if bits else "") + body).strip()
```

```python
# ---------------------------
# 8) 실행 파트 (NO-DF 모드): 단일 (SUBJECT_ID, HADM_ID, note_id) 대상으로 생성
# ---------------------------

# (사용자 입력부) 원하는 대상 설정
TOPK = 6
FETCH_K = 64
TARGET_SUBJECT_ID = 10000032
TARGET_HADM_ID    = 22595853
TARGET_NOTE_ID    = "10000032-DS-21"

# 검색 쿼리의 씨앗이 될 텍스트(요약/핵심 키워드 몇 줄이면 충분)
SEED_TEXT = (
    "Discharge summary seed. Focus on hospital course and discharge instructions. "
    "Include salient clinical facts that match the index."
)

# case_id 메타가 벡터DB 문서에 제대로 들어있다면 True로 사용하세요.
# 없다면 False로 두면 타 케이스도 허용하며 상위 유사도 결과에서 MMR로 고릅니다.
STRICT_MATCH = False

# 저장 경로
SAVE_PATH = f"/content/drive/MyDrive/DILAB/OK_Data/generated_{TARGET_SUBJECT_ID}_{TARGET_HADM_ID}_{TARGET_NOTE_ID}_noDF.parquet"

# (중요) df 없이 필요한 전역만 검사
for _n in ["vectorstore","emb_lc","tok","model","make_case_id","row_to_evidence_text"]:
    _require(_n)

# df 없이 실행할 row 딕셔너리 구성 (본문 컬럼을 하나 반드시 넣어야 함: TEXT/text/note_text/note/content/CONTENT/body 중 하나)
row = {
    "SUBJECT_ID": TARGET_SUBJECT_ID,
    "HADM_ID": TARGET_HADM_ID,
    "note_id": TARGET_NOTE_ID,
    "TEXT": SEED_TEXT,                    # row_to_evidence_text()가 자동 탐지할 본문
    "CATEGORY": "Discharge summary",      # 선택 필드(있으면 검색 품질 도움)
    "TITLE": "Discharge Note",            # 선택 필드(있으면 검색 품질 도움)
}

# 실행
try:
    bhc, di, used = generate_bhc_di_from_row(row, k=TOPK, fetch_k=FETCH_K, strict=STRICT_MATCH)
except Exception as e:
    bhc, di, used = f"ERROR: {e}", "Not documented", []

# 출력 미리보기
from datetime import datetime
print(f"[{datetime.now().strftime('%H:%M:%S')}] Completed 1/1 rows")
try:
    cid = make_case_id(row)
except Exception:
    cid = "N/A"
print(f"- case_id: {cid}  SUBJECT_ID: {TARGET_SUBJECT_ID}  HADM_ID: {TARGET_HADM_ID}  note_id: {TARGET_NOTE_ID}")
print(f"- BHC preview: {str(bhc)[:550].replace('\\n',' ')}")
print(f"- DI  preview: {str(di)[:550].replace('\\n',' ')}")

# 저장
import pandas as pd, os
os.makedirs(os.path.dirname(SAVE_PATH), exist_ok=True)
pd.DataFrame([{
    "SUBJECT_ID": TARGET_SUBJECT_ID,
    "HADM_ID": TARGET_HADM_ID,
    "note_id": TARGET_NOTE_ID,
    "bhc_gen": bhc,
    "di_gen": di,
    "ctx_doc_ids": used,
}]).to_parquet(SAVE_PATH, engine="pyarrow", compression="snappy", index=False)
print("\nSaved:", SAVE_PATH)
```
