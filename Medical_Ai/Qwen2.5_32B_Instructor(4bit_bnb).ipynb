```python
from huggingface_hub import snapshot_download
import shutil
import os

# 1. 경로 설정
# Colab 내부의 임시 고속 디스크 경로
local_temp_path = "/content/temp_qwen_download"
# 최종 저장될 구글 드라이브 경로
final_drive_path = "/content/drive/MyDrive/DILAB/Models/Qwen2.5-32B-bnb-4bit"
repo_id = "unsloth/Qwen2.5-32B-Instruct-bnb-4bit"

print(f"🚀 1단계: Colab 고속 디스크에 다운로드 시작 (속도 매우 빠름)")
# 먼저 로컬에 다운로드 (오류 확률 0에 가까움)
downloaded_path = snapshot_download(
    repo_id=repo_id,
    local_dir=local_temp_path,
    local_dir_use_symlinks=False,
    resume_download=True
)

print(f"\n✅ 다운로드 완료! 이제 구글 드라이브로 옮깁니다... (약 5~10분 소요)")

# 2. 구글 드라이브로 이동 (복사)
if not os.path.exists(final_drive_path):
    os.makedirs(final_drive_path, exist_ok=True)

# shutil을 이용해 안전하게 이동
import subprocess
# 리눅스 cp 명령어가 파이썬 shutil보다 훨씬 빠르고 안정적입니다.
command = f"cp -r -v '{local_temp_path}/.' '{final_drive_path}/'"
subprocess.run(command, shell=True)

print(f"\n🎉 모든 과정 완료! 모델이 구글 드라이브에 안전하게 저장되었습니다.")
print(f"저장 위치: {final_drive_path}")
```

```python
from unsloth import FastLanguageModel
import torch

# 방금 저장한 구글 드라이브 경로
model_path = "/content/drive/MyDrive/DILAB/Models/Qwen2.5-32B-bnb-4bit"

print("모델 로딩 중... (약 1~2분 소요)")

# 모델과 토크나이저 로드
model, tokenizer = FastLanguageModel.from_pretrained(
    model_name = model_path, # 인터넷 다운로드 없이 내 드라이브에서 로드!
    max_seq_length = 4096,
    dtype = None,
    load_in_4bit = True,
)

# 추론 테스트 (의학 질문)
messages = [
    {"role": "system", "content": "You are Medical assistant, please answer to Korean."},
    {"role": "user", "content": "응급의학과에서 퇴실지침서 작성하려면 어떤 정보가 필요할까?"}
]

# 포맷팅 및 생성
inputs = tokenizer.apply_chat_template(
    messages,
    tokenize = True,
    add_generation_prompt = True,
    return_tensors = "pt",
).to("cuda")

print("\n--- [Qwen 2.5 32B의 답변] ---")
outputs = model.generate(inputs, max_new_tokens = 512, use_cache = True)
print(tokenizer.batch_decode(outputs, skip_special_tokens=True)[0])
```
