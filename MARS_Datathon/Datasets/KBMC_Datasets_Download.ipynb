```python
from huggingface_hub import snapshot_download

# 저장하고 싶은 폴더 지정
SAVE_DIR = "/content/drive/MyDrive/DILAB/OK/DI_LAB/MARS_Datathon/Datasets/"

# 공개 데이터면 토큰 불필요
local_path = snapshot_download(
    repo_id="sungJoo/KBMC",
    repo_type="dataset",
    local_dir=SAVE_DIR,
    local_dir_use_symlinks=False,   # 캐시 심링크 말고 실제 파일 복사
    revision="main"                 # 필요 시 특정 커밋/태그로 고정 가능
)

print("Donwloaded to: ", local_path)
```

```python
import os, glob
ROOT = "/content/drive/MyDrive/DILAB/OK/DI_LAB/MARS_Datathon/Datasets"
print([p for p in glob.glob(os.path.join(ROOT, "KBMC*"), recursive=False)])
print([p for p in glob.glob(os.path.join(ROOT, "**/KBMC.tsv"), recursive=True)][:5])
```

```python
import os, shutil

SRC = "/content/drive/MyDrive/DILAB/OK/DI_LAB/MARS_Datathon/Datasets"
DST = os.path.join(SRC, "KBMC")
os.makedirs(DST, exist_ok=True)

for name in ["KBMC.tsv", "README.md", ".gitattributes"]:
    src = os.path.join(SRC, name)
    if os.path.exists(src):
        shutil.move(src, os.path.join(DST, name))

print("Moved to:", DST)
print("Now in DST:", os.listdir(DST))
```
