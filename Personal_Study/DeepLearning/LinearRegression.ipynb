## 선형회귀

```python
import numpy as np
import matplotlib.pyplot as plt

# 공부시간=x, 성적=y
x = np.array([2, 4, 6, 8])
y = np.array([81, 93, 91, 97])

# 데이터 분포를 그래프로 나타낸다
plt.scatter(x, y)
plt.show()

# 기울기 a의 값과 절편 b의 값을 초기화
a = 0
b = 0

# 학습률 설정
lr = 0.03

# 몇번 반복될지 설정
epochs = 2001

# x 값이 총 몇 개인지 세기
n = len(x)

# 경사 하강법
for i in range(epochs):
  y_pred = a * x + b
  error = y - y_pred

  a_diff = (2/n) * sum(-x * (error)) # 오차 함수를 a로 편미분
  b_diff = (2/n) * sum(-(error))     # 오차 함수를 b로 편미분
  a = a - lr * a_diff   # 학습률을 곱해 기존 a의 값을 업데이트
  b = b - lr * b_diff   # 학습률을 곱해 기존 b의 값을 업데이트

  if i % 100 == 0:    # 100번 반복될 때마다 현재의 a, b값을 출력
    print("epoch=%.f, 기울기=%.04f, 절편=%.04f" % (i, a, b))

# 앞서 구한 최종 a값을 기울기, b값을 y절편에 대입해 그래프 그리기
y_pred = a* x + b

plt.scatter(x, y)
plt.plot(x, y_pred, 'r')
plt.show()

```

## 다중 선형 회귀

```python
# 공부시간=x, 성적=y
x1 = np.array([2, 4, 6, 8])
x2 = np.array([0, 4, 2, 3])
y = np.array([81, 93, 91, 97])
```

```python
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.scatter(x1, x2, y)
plt.show()
```

```python
a1= 0
a2 = 0
b = 0
lr = 0.01
epochs = 2001

n = len(x1)

for i in range(epochs):
  y_pred = a1*x1 + a2*x2 + b
  error = y - y_pred

  a1_diff = (2/n) * sum(-x1*(error))
  a2_diff = (2/n) * sum(-x2*(error))
  b_diff = (2/n) * sum(-(error))

  a1 = a1 -lr * a1_diff
  a2 = a2 - lr*a2_diff
  b = b - lr * b_diff

  if i % 100 == 0:
    print("peoch=%.f, 기울기=%.04f, 기울기2=%.04f, 절편=%.04f" % (i, a1, a2, b))

  print("실제 점수: ", y)
  print("예측 점수: ", y_pred)
```
